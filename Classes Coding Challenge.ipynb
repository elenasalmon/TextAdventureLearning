```python
import numpy as np
import numpy.random as random
```

```python
def roll(nsides):
    return random.randint(nsides) + 1

class Dice():
    
    def __init__(self, isides):
        self.sides = isides
        pass
    
    def roll_dice(self):
        return roll(self.sides)
    
class ManyDice(Dice):
    
    def __init__(self, inum, isides):
        super().__init__(isides)
        self.num = inum
        pass
    
    def roll_multiple_dice(self):
        r = []
        for i in range(self.num):
            r_ = roll(self.sides)
            r.append(r_)
        return sum(r)
```

```python
def defence(monster_attack, defence_stat, player_hp):
    damage = monster_attack - defence_stat #monster_attack is an input of damage
    if damage > 0:
        player_hp = player_hp - damage
        #if player_hp >0: #should add deus ex machina to make sure you stay above zero hp, maybe here like this?
        print("You were hit! You take", damage, "damage! You have", player_hp, "health points remaining.")
        return player_hp        
            
    else:
        print("You deflected the attack! You take no damage, and still have", player_hp, "health points remaining.")
        return player_hp
```

```python
def attack_action(attack_damage, monster_defence, monster_hp):
    damage = attack_damage - monster_defence
    if damage > 0:
        monster_hp = monster_hp - damage
        if monster_hp <= 0:
            print("You dealt the winning blow! You have defeated your inner demon!")
        else:
            print("You did", damage, "points of damage. Your inner demon still has", monster_hp, "health points remaining.")
            return monster_hp
    else:
        print("The inner demon deflected your attack! It still has", monster_hp, "health points remaining.")
        return monster_hp
```

```python
class PlayerCharacter():
    
    def __init__(self):
        #starting hp
        self.hp = 5 #30
        #starting stats for the different attacks (you start with a d6)
        self.sl = 6 #self-love
        self.sk = 6 #self-kindness
        self.k = 6 #kindness
        self.c = 6 #compassion
        #defence starting stats - will be overridden when choose player
        #class, but inputted here so can define defence
        #maybe if no class chosen, default to "Stable" which is:
        self.future = 5
        self.present = 5
        self.past = 5
        pass
```

```python
class Anxious(PlayerCharacter):
    
    def __init__(self):
        super().__init__()
        # defence stats:
        self.future = 2
        self.present = self.future
        self.past = 5
```

```python
class LowMood(PlayerCharacter):
    
    def __init__(self):
        super().__init__()
        # defence stats:
        self.future = 5
        self.past = 2
        self.present = self.past
```

```python
class Stable(PlayerCharacter):
    
    def __init__(self):
        super().__init__()
        # defence stats:
        self.future = 5
        self.present = 5
        self.past = 5
```

```python
class AnxiousAndLowMood(PlayerCharacter):
    
    def __init__(self):
        super().__init__()
        # defence stats:
        self.future = 2
        self.present = (self.past + self.future)/2
        self.past = 2
```

```python
def attack_roll(attack_stat):
    d6 = Dice(6)
    d12 = Dice(12)
    twod6 = ManyDice(2, 6)
    twod12 = ManyDice(2, 12)
    if attack_stat == 6:
        return d6.roll_dice()
    if attack_stat == 12:
        return d12.roll_dice()
    if attack_stat == {2, 6}:
        return twod6.roll_multiple_dice()
    if attack_stat == {2, 12}:
        return twod12.roll_multiple_dice()
    else:
        print("Error: invalid attack stat/dice inputted. Please report this bug to Elena Salmon - thank you.")
```

```python
def game():
    print("Welcome to this game of daily life!")
    
    #sequence to start the game - added so that can have starting explanation sentence
    while True:
        start = start_game()
        if start == None:
            print("Type 'Yes' to Continue:")
        else:
            break
    
    #player to choose game difficulty
    while True:
        monster = set_difficulty()
        if monster == None:
            print("Please try again:")
        else:
            break
    
    #player to choose their class (mental health)
    while True:
        player = choose_class()
        if player == None:
            print("Please try again:")
        else:
            break
            
    print(f"You will start with {player.hp} health points, and your inner demon will start with {monster.hp} health points. You will win if your inner demon is reduced to 0 health points.") 
    #start the game proper with attacks and defence
    
    global GameOver
    GameOver = False
    while GameOver == False:
            
        if monster.hp <= 0:
            GameOver = True
            break
            
        if player.hp <= 0: # death condition - hopefully never to be used in final game
            print("You died. Game Over.")
            GameOver = True
            break
            
        #monster attacks player
        if GameOver == False:
            x = roll(3)
            if x == 1: #future-type attack
                print("Your inner monologue starts to think about a negative thing you have to do in the future")
                player.hp = defence(attack_roll(monster.future), player.future, player.hp)
            if x == 2: #present-type attack
                print("Your attention starts to zone in on a negative thing happening in front of you")
                player.hp = defence(attack_roll(monster.present), player.present, player.hp)
            if x == 3: #past-type attack
                print("Your inner monologue starts to ruminate on a negative moment from your past")
                player.hp = defence(attack_roll(monster.past), player.past, player.hp)
        else:
            break
        
        #player attacks monster
        if GameOver == False:
            stat = choose_attack(player, monster)
            if stat == None:
                print("Please try again:")
            else:
                monster.hp = stat
        else:
            break
```
